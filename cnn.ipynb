```python
from sklearn.model_selection import train_test_split
```

```python
X =  df.drop(['label'],axis='columns')
```

```python
y = df['label']
```

```python
y.head()
```

```python
myDict={'A':1,'B':2,'C':3,'D':4,'E':5,'F':6,'G':7,'H':8,'I':9,'K':10,'L':11,'M':12,'N':13,'O':14,'P':15,'Q':16,'R':17,'S':18,'T':19,'U':20,'V':21,'W':22,'X':23,'Y':24}
```

```python
for i in range(len(y)):
    y.iloc[i] = myDict[y.iloc[i]]
```

```python
y
```

```python
y = y.astype(int)
y
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
len(X_train)
```

```python
y_train
```

```python
len(X_test)
```

```python
X_train.shape
```

```python
from tensorflow import keras
```

```python
model1 = keras.Sequential([
keras.layers.Dense(100,input_shape=(30000,),activation= 'relu'),
keras.layers.Dense(100,input_shape=(30000,),activation= 'relu'),
keras.layers.Dense(100,input_shape=(30000,),activation= 'relu'),
keras.layers.Dense(100,input_shape=(30000,),activation= 'relu'),
keras.layers.Dense(100,input_shape=(30000,),activation= 'relu'),
keras.layers.Dense(30,activation= 'softmax')])
# keras.add(Dense(7, activation='softmax'))
#complie model
model1.compile(loss='sparse_categorical_crossentropy', optimizer= 'adam', metrics=['accuracy'])
```

```python
y_train[0:100]
```

```python
model1.fit(X_train,y_train,epochs=15)
```

```python
y_test
```

```python
 model1.evaluate(X_test,y_test)
```

```python
import numpy as np
y_predicted=model1.predict(X_test)
y_predicted_labels = [np.argmax(i) for i in y_predicted]
```

```python
import tensorflow as tf
cm=tf.math.confusion_matrix(labels=y_test,predictions=y_predicted_labels)
cm
```

```python
import seaborn as sn 
plt.figure(figsize=(20,10))
sn.heatmap(cm,annot=True,fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
